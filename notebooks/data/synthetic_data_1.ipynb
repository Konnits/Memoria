# Datos sintéticos

Crearemos datos sintéticos no espaciados para poder entrenar el Transformer.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

FEATURES = 1
N_SAMPLES = 1_000_000
MIN_TIME = 0
MAX_TIME = 10_000_000

time = np.random.uniform(MIN_TIME, MAX_TIME, N_SAMPLES)

def synthetic_function(t):
    return (0.01 * np.sin(0.00001 * t) ** 2 + 0.05 * np.sin(0.0001 * t) ** 2 + 0.02 * np.sin(0.001 * t) ** 2) * np.sqrt(t) + np.random.normal(0, 1, t.shape)

df = pd.DataFrame({
    'timestamp': time,
    'value': synthetic_function(time)
})
```

```python
df.sort_values('timestamp', inplace=True)
df.reset_index(drop=True, inplace=True)
df.to_csv('../../data/processed/synthetic_data_1.csv', index=False)
```
